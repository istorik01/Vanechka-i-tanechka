```python
def movie_runtime_minutes(x):
    if pd.isna(x):  
        return None
    
    numbers = re.findall(r'[0-9]+', str(x))
    
    if len(numbers) == 0:  
        return None
    elif len(numbers) == 1:  
        return int(numbers[0]) * 60
    else:  
        return int(numbers[0]) * 60 + int(numbers[1])
```

```python
df['Movie Runtime Minutes'] = df['Movie Runtime'].apply(movie_runtime_minutes)
```
